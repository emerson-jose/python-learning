```python
# lista inicial marcas de carros

marcas_carro = ["FORD", "GM", "BYD", "HONDA"]
# adicionar toyota ao final da lisca com o append
marcas_carro.append("TOYOTA")

#imprimir a lista apos a adição

print("Lista após adicionar Toyota:", marcas_carro )

#remoção do elemento  da posição 1 ("GM") utilizando pop

marcas_carro.pop(1)

#impreção da lista apos a remoção

print("Lista após remover GM:",marcas_carro)
```

```python
# Cria a lista usando range de 2 a 150 (inclusive)
lista = list(range(2, 151))

# Imprime todos os elementos da lista
print("Lista completa:")
print(lista)

# Usa fatiamento para imprimir os seis números entre as posições 75 e 80 (índices 75 a 80)
print("\nElementos nas posições 75 a 80:")
print(lista[75:81])
```

```python
# Criação do dicionário estudante
estudante = {
    "nome": "Carlos",
    "idade": 22,
    "nota": 7.8
}

# Impressão do valor da chave "nota"
print("Nota do estudante:", estudante["nota"])
```

```python
# Adiciona a chave "disciplina" com o valor "Matemática"
estudante["disciplina"] = "Matemática"

# Altera o valor da chave "idade" para 23
estudante["idade"] = 23

# Imprime o dicionário atualizado

print(estudante)
```

```python
# Armazena o número na variável 'valor'
valor=7

# Verifica as condições usando if, elif e else
if valor > 0:
    print("Positivo")
elif valor < 0:
    print("Negativo")
else:
    print("Zero")
```

```python
# Armazena o número na variável
valor = 21

# Verifica se o número é divisível por 3
if valor  %3 == 0:
    print("Divisível por 3")
else:
    print("Não divisível por 3")
```

```python
def media_lista(lista):
    # Calculo da média dividindo a soma dos elementos pela quantidade de elementos
    return sum(lista) / len(lista)

# Teste da função com a lista [10, 20, 30, 40]
print(media_lista([10, 20, 30, 40]))
```
