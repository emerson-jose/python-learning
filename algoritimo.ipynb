```python
# Exemplos de valores booleanos
verdadeiro = True
falso = False

print(f"Vari√°vel 'verdadeiro': {verdadeiro}")
print(f"Vari√°vel 'falso': {falso}")

print("\n--- Operadores de Compara√ß√£o ---")
# Operadores de compara√ß√£o retornam True ou False
x = 10
y = 5

print(f"{x} > {y}: {x > y}")   # Maior que
print(f"{x} < {y}: {x < y}")   # Menor que
print(f"{x} == {y}: {x == y}") # Igual a
print(f"{x} != {y}: {x != y}") # Diferente de
print(f"{x} >= {y}: {x >= y}") # Maior ou igual a
print(f"{x} <= {y}: {x <= y}") # Menor ou igual a

print("\n--- Operadores L√≥gicos ---")
# Operadores l√≥gicos combinam valores booleanos
p = True
q = False

print(f"{p} and {q}: {p and q}") # AND l√≥gico (ambos True)
print(f"{p} or {q}: {p or q}")   # OR l√≥gico (pelo menos um True)
print(f"not {p}: {not p}")      # NOT l√≥gico (inverte o valor)
```

```python
while True:
    seu_nome = input("digite seu nome:")
    peso_str = input("digite seu peso:")
    altura_str = input("digite a sua altura:")

    try:
        # Replace comma with period for decimal conversion and convert to float
        peso = float(peso_str.replace(',', '.'))
        altura = float(altura_str.replace(',', '.'))

        # IMC calculation: weight / (height in meters)^2
        IMC = peso / (altura ** 2)

        print(f"Ol√° {seu_nome} seu IMC foi de {IMC:.2f} ")

        if IMC < 18.5:
            print(f'{seu_nome}, seu peso est√° abaixo do normal.')
        elif IMC >= 18.5 and IMC <= 24.9:
            print(f'{seu_nome}, seu peso est√° normal.')
        else:
            print(f'{seu_nome}, voc√™ est√° acima do peso.')

    except ValueError:
        print("Entrada inv√°lida. Por favor, digite n√∫meros para peso e altura.")
    except ZeroDivisionError:
        print("Altura inv√°lida. A altura n√£o pode ser zero.")
```

```python
while True:
    nota1_str = input("digite a primeira nota:")
    nota2_str = input('digite a segunda nota:')

    try:
        # Replace comma with period for decimal conversion and convert to float
        nota1 = float(nota1_str.replace(',', '.'))
        nota2 = float(nota2_str.replace(',', '.'))

        soma = nota1 + nota2
        media = soma / 2

        print(f"A m√©dia das notas √©: {media:.2f}")

        if media >= 7:
            print("aprovado")
        elif media >= 5: # Use >= 5 for recovery, assuming less than 7
            print("voce esta de recupera√ß√£o!")
        else:
            print("vc esta reprovado")

    except ValueError:
        print("Entrada inv√°lida. Por favor, digite n√∫meros para as notas.")
    except ZeroDivisionError:
        # This exception is less likely here unless input handling allows it
        print("Erro na divis√£o. Verifique as notas inseridas.")
```
