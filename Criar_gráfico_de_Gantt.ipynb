```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.patches as mpatches

```

```python
# Linha do tempo gerada pela simulação
linha_tempo = [
    'P1', 'P1', 'P2', 'P2', 'P3', 'P3', 'P4', 'P4', 'P1', 'P1', 'P2', 'P2', 'P3', 'P3',
    'P1', 'P4', 'P4', 'P4', 'P4', 'P2', 'P2', 'P4', 'P4', 'P2', 'P2', 'P3', 'P3', 'P4',
    'P4', 'P1', 'P1', 'P2', 'P2', 'P3', 'P3', 'P4', 'P4', 'P1', 'P1', 'P2', 'P3', 'P3',
    'P1', 'P1', 'P3', 'P3', 'P1', 'P1', 'P3', 'P1', 'P1', 'P1'
]
```

```python
# Cores por processo
cores = {
    "P1": "skyblue",
    "P2": "lightgreen",
    "P3": "salmon",
    "P4": "plum",
    "IDLE": "lightgray"
}
```

```python
# Criar gráfico de Gantt
fig, ax = plt.subplots(figsize=(15, 2))

for t, processo in enumerate(linha_tempo):
    ax.broken_barh([(t, 1)], (0, 5), facecolors=cores.get(processo, "black"))

ax.set_ylim(0, 5)
ax.set_xlim(0, len(linha_tempo))
ax.set_xlabel("Tempo (UT)")
ax.set_yticks([])
ax.set_title("Gráfico de Gantt - Escalonador Round-Robin (Quantum = 2 UT)")
```

```python
# Legenda
legenda = [mpatches.Patch(color=cor, label=proc) for proc, cor in cores.items()]
ax.legend(handles=legenda, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
